# What makes a Movie Great?

Theyre are many things that can define a movie as great. Popularity, box office revenue, voter average, even the impact it had on the viewers. While everyone is entitled to their own opinion and lists, these things cannot take away from what one movie goers thinks opposed to another.

```python
movieRevenue = newMovies.sort_values(by="revenue", ascending=False).head(501)
```

```python
plt.figure(figsize = (15, 10))
plt.bar("title", "revenue", data = movieRevenue, color= 'green')
plt.xlabel('Movie name')
plt.xticks(rotation = 90, fontsize = 'large')
plt.ylabel('revenue in millions')
plt.title('Movie Revenue')
plt.show()
```

```python
movieRevenue.head(51)
```

I first wanted to look as revenue, supply and demand play a big factor into the way our global economy works and the same works for movies. However different it may be the basic principle of supply and demand still applies to movies. As you can see here in our revised dataframe of 500 movies, in the top ten most successful movies, we have 3 movies that sit under the 'Adventure" id followed but 2 for 'Action' and 'comedy' but i also wanted to note that a lot of movies here dont have little to no profit. This shows that movies can and have lost profit

From this alone it can seem that viewers enjoy a movie that follows a main protagonist on his journey that also includes risk and danger. However I want to take a look at voter average and popularity to see if this remains true

```python
moviePopularity = newMovies.sort_values(by="popularity", ascending=False).head(51)
```

```python
plt.figure(figsize = (15, 5))
plt.xticks(rotation = 90, fontsize = 'large')
sns.barplot(x='title', data = moviePopularity,  y = 'popularity',)
```

```python
moviePopularity.head(10)
```

Here we have the the 50 most popular movies according to the list. Now the interesting part is that the average movie has a popularity of around 20. the 4 movies you see break that trend are Pulp Fiction, Shawshank, Forrest Gump, and Star Wars. the difference between these 4 and the rest is that for the time of their release. These were culture breaking movies, these 4 took the US by storm and even to this day remain just as popular. 

```python
movieAvg = newMovies.sort_values(by="vote_average", ascending=False).head(51)
```

```python
plt.figure(figsize = (15, 5))
plt.xticks(rotation = 90, fontsize = 'large')
sns.barplot(x='title', data = movieAvg,  y = 'vote_average',)
```

```python
movieCount = newMovies.sort_values(by="vote_count", ascending=False).head(51)
```

```python
plt.figure(figsize = (15, 5))
plt.xticks(rotation = 90, fontsize = 'large')
sns.barplot(x='title', data = movieCount,  y = 'vote_count',)
```

Here, we have "vote_average" and "vote_count", with this data it is easy to see that Revenue, popularity, and Vote_average all correlate with movie success. but as shown with vote average, the data can be mislead as the movie "Girl in the Cadillac" and "Reckless" werent even in the top 50 for fan count
